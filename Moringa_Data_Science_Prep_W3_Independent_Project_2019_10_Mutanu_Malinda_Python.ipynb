```
# Loading the data
cdr2 = pd.read_csv('http://bit.ly/Telcom_dataset2')

cdr2.head(10)
```

```
# Describing the dataframe
cdr2.info()

cdr2.describe()
```

```
# Standardizing column names
cdr2.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

cdr2.head(5)
```

```
# Loading the data
cdr3 = pd.read_csv('http://bit.ly/Telcom_dataset3')

cdr3.head(10)
```

```
# Describing the dataframe
cdr3.info()

cdr3.describe()
```

```
# Standardizing column names
cdr3.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

cdr3.head(5)
```

```
# Merging the dataframes
cdr = pd.concat([cdr1, cdr2, cdr3], axis=0)
cdr
```

```
# Converting value to int data type
cdr['VALUE'] = pd.to_numeric(cdr['VALUE'])

cdr.info()
```

```
# Deleting unnecessary columns
cdr = cdr.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)

cdr.head(10)
```

```
#Viewing duplicates
cdr[cdr.duplicated()].sort_values(by = 'SITE_ID')
```

```
# Checking for missing values
cdr.isnull().sum()
```

```
# Dropping records where value is 0
cdr = cdr[cdr.VALUE != 0]
cdr
```

```
# Finding the services that return the lowest value
a = cdr.VALUE.groupby(cdr['PRODUCT']).mean()

a.sort_values()
```

```
# Merging the dataframes on cell ID
df3 = pd.merge(df2, cdr, how='outer', on=('CELL_ID'))
df3
```

```
# Identifying regions with status as "Not In Service"

b = df3[df3['STATUS'] == "Not In Service"].groupby(df3['REGION']).count()
b.sort_values()[:5]
```

```
# Identifying regions returning the highest and lowest value

df3[df3['VALUE']].groupby(df3['REGION']).sort_values(ascending= 0)[:5]
```
