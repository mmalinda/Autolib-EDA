```
# Loading packages for exploraion and analysis
import pandas as pd
import numpy as np
```

## cells_geo dataset

```
# Loading the cells_geo dataframe
cells_geo = pd.read_csv('cells_geo.csv')

cells_geo
```

```
# Describing the dataframe
cells_geo.info()

cells_geo.describe()
```

```
# Splitting the data into columns
df = pd.DataFrame(cells_geo[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].str.split(';').tolist())

df
```

```
# Renaming columns
df.columns = ['index', 'VILLES', 'STATUS', 'LOCALISATION', 'DECOUPZONE', 'ZONENAME', 'LONGITUDE', 'LATITUDE', 'REGION', 'AREA', 'CELL_ID', 'SITE_CODE']

df
```

```
# Describing the dataframe
df.info()
df.describe()
```

```
# Converting longitude and latitude to float data type
df['LATITUDE'] = pd.to_numeric(df['LATITUDE'])
df['LONGITUDE'] = pd.to_numeric(df['LONGITUDE'])

df.info()
```

```
# Removing extra characters from decoupzone
df['DECOUPZONE'] = df['DECOUPZONE'].str.replace('"', "")
df
```

```
# Finding missing values
df = df.replace("", np.nan)
df
```

```
#Finding out how many records have missing values
df.isnull().sum()
```

```
# Assessing missing values
df.head(60)
```

```
#Dropping rows with missing values
df2 = df.dropna(subset = ['STATUS'], axis = 0)
df2
```

```
# Searching for duplicates in the data
df2[df2.duplicated()]
```

# cdr Dataset

```
# Loading the data
cdr1 = pd.read_csv('http://bit.ly/Telcom_dataset1')

cdr1.head(10)
```

```
# Describing the dataframe
cdr1.info()

cdr1.describe()
```

```
# Standardizing column names
cdr1.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

cdr1.head(5)
```

```
# Loading the data
cdr2 = pd.read_csv('http://bit.ly/Telcom_dataset2')

cdr2.head(10)
```

```
# Describing the dataframe
cdr2.info()

cdr2.describe()
```

```
# Standardizing column names
cdr2.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

cdr2.head(5)
```

```
# Loading the data
cdr3 = pd.read_csv('http://bit.ly/Telcom_dataset3')

cdr3.head(10)
```

```
# Describing the dataframe
cdr3.info()

cdr3.describe()
```

```
# Standardizing column names
cdr3.columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

cdr3.head(5)
```

```
# Merging the dataframes
cdr = pd.concat([cdr1, cdr2, cdr3], axis=0)
cdr
```

```
# Converting value to int data type
cdr['VALUE'] = pd.to_numeric(cdr['VALUE'])

cdr.info()
```

```
# Deleting unnecessary columns
cdr = cdr.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)

cdr.head(10)
```

```
#Viewing duplicates
cdr[cdr.duplicated()].sort_values(by = 'CELL_ID')
```

```
# Checking for missing values
cdr.isnull().sum()
```
